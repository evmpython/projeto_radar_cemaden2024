Fun√ß√£o que extrai as informa√ß√µes do dado

```python
def get_info_rainbow_vol(filename):

    # importa bibliotecas
    import wradlib as wrl
    import pandas as pd

    # leitura do radar
    fcontent = wrl.io.read_rainbow(filename)

    # informa√ß√µes do radar
    version = fcontent["volume"]["@version"]
    datetime = fcontent["volume"]["@datetime"]
    type = fcontent["volume"]["@type"]
    
    # valores do 'sensorinfo'
    sensor_info = fcontent["volume"]["sensorinfo"]
    sensor_info = pd.DataFrame({'info': sensor_info.keys(), 'valores': sensor_info.values()})
    
    # extrai as informa√ß√µes
    elevation, range_step, high_prf, start_range, stop_range, azimuths, bins = [], [], [], [], [], [], []
    nelev = 10
    for ielev in range(nelev):
        elevation.append(fcontent["volume"]["scan"]['slice'][ielev]['posangle'])
        range_step.append(fcontent["volume"]["scan"]['slice'][ielev]['rangestep'])
        high_prf.append(fcontent["volume"]["scan"]['slice'][ielev]['highprf'])
        start_range.append(fcontent["volume"]["scan"]['slice'][ielev]['start_range'])
        stop_range.append(fcontent["volume"]["scan"]['slice'][ielev]['stoprange'])
        azimuths.append(fcontent["volume"]["scan"]['slice'][ielev]['slicedata']['rawdata']['@rays'])
        bins.append(fcontent["volume"]["scan"]['slice'][ielev]['slicedata']['rawdata']['@bins'])

    # insere num dicion√°rio
    dic = {'version': version,
           'datetime': datetime,
           'type': type,

           'name_radar': fcontent["volume"]["sensorinfo"]['@name'],
           'lat_radar': fcontent["volume"]["sensorinfo"]['lat'],
           'lon_radar': fcontent["volume"]["sensorinfo"]['lon'],
           'alt_radar': fcontent["volume"]["sensorinfo"]['alt'],
           'wavelen': fcontent["volume"]["sensorinfo"]['wavelen'],
           'beamwidth': fcontent["volume"]["sensorinfo"]['beamwidth'],
           
           'elevations': elevation,
           'range_step': range_step,
           'high_prf': high_prf,
           'start_range': start_range,
           'stop_range': stop_range,
           'azimuths': azimuths,
           'bins': bins}
    dic = pd.DataFrame({'info': dic.keys(), 'valores': dic.values()})

    dic.to_csv(f'radar_info_{fcontent["volume"]["sensorinfo"]["@name"]}.csv', index=False, header=False)

    return dic

# chama fun√ß√£o
filename = '2024010300113700dBZ.vol'
get_info_rainbow_vol(filename)
```
